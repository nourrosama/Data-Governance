# **Data Security**
-Anonymization: It is the process of irreversibly transforming data in such a way that it can no longer be associated with an individual. The goal is to render the data anonymous, meaning that even if someone gains access to the data, they cannot identify the individuals to whom the data belongs.

-Pseudonymization: It involves replacing direct identifiers in data with artificial identifiers, known as pseudonyms. Unlike anonymization, pseudonymization allows for the possibility of re-identification but only by authorized parties who possess the corresponding pseudonymization key.

**Data anonymization**

1.Randomization:

Randomization involves replacing sensitive data with random values. For example, a person's name might be replaced with a randomly generated string of characters.

```python
!pip install faker
```

```python
import string
import random

# define function that operates on entire dataframe
def randomize_values(df):
    for column in df.columns:
        if df[column].dtype == 'O': # check if column has object dtype
            df[column] = [''.join(random.choices(string.ascii_letters + string.digits, k=10))
            for _ in range(len(df))] # generate a list of random strings
    return df
```

```python
# apply function to dataframe
df_rand1 = randomize_values(movies)

df_rand1
```

1.Generalization:

Generalization involves replacing specific values with more general or less precise values. It involves transforming or categorizing data into broader categories or ranges. For example, instead of storing exact ages, you might generalize ages into age groups (e.g., 20-30, 30-40, etc.).

```python
# Anonymize the votes by grouping them into votes ranges
bins = [150, 350, 500, 2000, 8000, 15000, 50000, 100000, 500000, 900000, 1000000]
labels = ['0-149', '150-349', '350-499', '500-2k', '2k-8k', '8k-15k', '15k-50k', '50k-100k', '100k-500k', '500k-900k']
movies['VOTES'] = pd.cut(movies['VOTES'], bins=bins, labels=labels)



# Anonymize the runtime by grouping them into ranges
bins = [15, 35, 50, 120, 150, 180]
labels = ['0-15 min', '15 min - 35 min', '35 min - 50 min', '50 min -120 min', '120 min - 150 min']
movies['RunTime'] = pd.cut(movies['RunTime'], bins=bins, labels=labels)

# Anonymize the rating by grouping them into ranges
bins = [4, 5, 6, 7, 8, 9]
labels = ['Very Low', 'Low', 'Average', 'Above Average', 'High']
movies['RATING'] = pd.cut(movies['RATING'], bins=bins, labels=labels)



movies
```

 **Role-Based Access Control (RBAC)**

RBAC is a widely used access control model that assigns permissions to users based on their roles within an organization. I have defined roles such as "admin," "data analyst," or "viewer" and assigned appropriate access permissions to each role. For example, an admin role may have full access to all columns, while a viewer role may only have read access to certain columns like MOVIES, YEAR, and GENRE.

```python
import pandas as pd

# Example RBAC data (modify as needed)
roles = ['admin', 'data analyst', 'editor', 'viewer']
permissions = ['read', 'create', 'update', 'delete']

# Create a DataFrame
rbac_df = pd.DataFrame(columns=['Role'] + permissions)
rbac_df['Role'] = roles
rbac_df.set_index('Role', inplace=True)

# Fill in permission values (e.g., 1 for granted, 0 for denied)
rbac_df.loc['admin', ['create', 'read', 'update', 'delete']] = 1
rbac_df.loc['data analyst', ['create', 'read', 'update']] = 1
rbac_df.loc['editor', ['update', 'delete']] = 1
rbac_df.loc['viewer',  'read'] = 1

print(rbac_df)
```

 **Attribute-Based Access Control (ABAC):**

ABAC provides access control based on the attributes or properties associated with data elements. In movie dataset, may define policies that allow or restrict access to specific columns based on attribute values. For instance, set a policy that allows access to the RUNTIME column only if the user has a certain level of authorization or belongs to a specific group.

```python
import pandas as pd

# Create a sample dataframe
data = {
    "user_id": [1, 2, 3],
    "name": ["Max", "Nina", "Alice"],
    "ip_address": ["127.0.0.1", "192.168.1.1", "10.0.0.1"]
}

users_df = pd.DataFrame(data)
print(users_df)
```

```python
!pip install py_abac
```

```python
from py_abac import PDP, Policy, AccessRequest
from py_abac.storage.memory import MemoryStorage

def check_access(user_df, user_name, action, client_ip):
    """
    Check if access is allowed based on user attributes.
    Args:
        user_df (pd.DataFrame): Users dataframe.
        user_name (str): User's name.
        action (str): Action (e.g., "create", "delete", "get").
        client_ip (str): Client IP address.

    Returns:
        bool: True if access is allowed, False otherwise.
    """
    allowed_users = ["Max", "Nina"]
    allowed_actions = ["create", "delete", "update"]

    if user_name in allowed_users and action in allowed_actions and client_ip == "127.0.0.1":
        return True
    else:
        return False

# Example access request
user_name = "Max"
action = "update"
client_ip = "127.0.0.1"

# Check if access is allowed
if check_access(users_df, user_name, action, client_ip):
    print(f"Access granted for user '{user_name}' to perform action '{action}'.")
else:
    print(f"Access denied for user '{user_name}' to perform action '{action}'.")

```

```python
# Create an in-memory policy
policy_json = {
    "uid": "1",
    "description": "Max and Nina are allowed to create, delete, get any resources only if the client IP matches.",
    "effect": "allow",
    "rules": {
        "subject": [
            {"$.name": {"condition": "Equals", "value": "Max"}},
            {"$.name": {"condition": "Equals", "value": "Nina"}}
        ],
        "resource": {"$.name": {"condition": "RegexMatch", "value": ".*"}},
        "action": [
            {"$.method": {"condition": "Equals", "value": "create"}},
            {"$.method": {"condition": "Equals", "value": "delete"}},
            {"$.method": {"condition": "Equals", "value": "update"}}
        ],
        "context": {"$.ip": {"condition": "CIDR", "value": "127.0.0.1/32"}}
    },
    "targets": {},
    "priority": 0
}

# Create a policy object
policy = Policy.from_json(policy_json)

# Set up in-memory policy storage
storage = MemoryStorage()
storage.add(policy)

# Create a Policy Decision Point (PDP)
pdp = PDP(storage)

# Example access request
request_json = {
    "subject": {"id": "", "attributes": {"name": "Habiba"}},
    "resource": {"id": "", "attributes": {"name": "myrn:example.com:resource:123"}},
    "action": {"id": "", "attributes": {"method": "update"}},
    "context": {"ip": "127.0.0.1"}
}

# Create an access request object
request = AccessRequest.from_json(request_json)

# Check if access request is allowed
if pdp.is_allowed(request):
    print(f"Access granted for user '{user_name}' to perform action '{action}'.")
else:
    print(f"Access denied for user '{user_name}' to perform action '{action}'.")
```
